```python
import pandas as pd                                   # library to work with data
```

```python
from matplotlib import pyplot as plt                  # library for creating visual representations of data
```

```python
heartData = pd.read_csv('Heart Attack Data Set.csv')  # load the csv file into a variable
```

```python
heartData                                             # use variable to display the data
```

```python
type(heartData)
```

```python
plt.plot(heartData.age,heartData.cp, 'ro')    # 'o' is used for a scatterplot and the letter before 'o' is the color
plt.title('heart attack data', fontsize=24, color='b')
plt.xlabel('age')
plt.ylabel('chest pain type')
```

```python
plt.bar(heartData.age,heartData.cp,color='y')
plt.title('heart attack data', fontsize=24, color='b')
plt.xlabel('age', color='b')
plt.ylabel('chest pain type', color='r')
```

```python
age_group = heartData.groupby('age')                 # group by the number of people in each age group
```

```python
age_group.size()                                     # display the size of each age group
```

```python
plt.plot(age_group.size(), 'o')
plt.title('age of participants', fontsize=24)                        # plot the size of each age group
plt.xlabel('age', color='b', fontsize=14)
plt.ylabel('number of people', fontsize=14, color='b')
```
